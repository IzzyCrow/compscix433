```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

#y = relative humidty now re-do assignment
```

```python
# Import the Excel as a pandas dataframe
air = pd.read_excel("AirQualityUCI.xlsx")
```

```python
# Report on the number of rows and columns in the dataframe
print (air.shape)
```

```python
# Print the top five rows of the dataframe
print(air.head)
```

```python
# Print the data types of the columns
print (air.dtypes)
```

```python
# fill the NaN values in each column for the column mean
air.fillna(air.mean())
```

```python
#  Find the Correlation

corr = air.corr()
corr.style.background_gradient(cmap='RdBu')
```

C6H6(GT) and AH are one of the least correlated variables


```python
linreg = linear_model.LinearRegression()
```

```python
# Create the dataseries of the least correlated fields for the model
air_ah = air['AH']
air_C6H6 = air['C6H6(GT)']

# Generate the Training Set
x_train, x_test, y_train, y_test = train_test_split(air_ah, air_C6H6, test_size=0.2,random_state=42)

# Reshape the Training Set
y_train = y_train[:,None]
x_train = x_train[:,None]

# Calculate the Regression
linreg.fit(x_train, y_train)
```

```python
# Predict Y

# Reshape the Test Set
y_test = y_test[:,None]
x_test = x_test[:,None]


yhat = linreg.predict(x_test)
```

```python
# The Mean Squared Error
print(np.mean((yhat-y_test)**2))

print((np.mean(yhat-y_test)**2)/np.mean(y_test))
```
